# Swarm plot

```python
read_file1 = pd.read_csv(r'state_wise_cases.csv')
df_11 = DataFrame(read_file1)
df_11.fillna(0)
```

```python
df_filtered = df_11[(df_11['Name of State / UT'] == 'Rajasthan') | (df_11['Name of State / UT'] == 'Kerala') | (df_11['Name of State / UT'] == 'Delhi') | (df_11['Name of State / UT'] == 'Tamil Nadu') | (df_11['Name of State / UT'] == 'Maharashtra') | (df_11['Name of State / UT'] == 'Andhra Pradesh') | (df_11['Name of State / UT'] == 'Uttar Pradesh')]
sb.swarmplot(x="Name of State / UT", y="Cured/Discharged/Migrated", data=df_filtered)
plt.xticks(rotation = 70)
plt.xlabel('Name of State / UT',size = 20)
plt.ylabel('Cured/Discharged/Migrated',size = 20)
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.show()
```

# Boxenplots

```python
read_file2 = pd.read_csv(r'state_wise_cases.csv')
df_12 = DataFrame(read_file2)
df_12.fillna(0)
```

```python
df_12.drop(df_12.index[0:135], inplace = True)
#df_12.reset_index(drop = True)
df_filtered = df_12[(df_12['Name of State / UT'] == 'Rajasthan') | (df_12['Name of State / UT'] == 'Kerala') | (df_12['Name of State / UT'] == 'Delhi') | (df_12['Name of State / UT'] == 'Tamil Nadu') | (df_12['Name of State / UT'] == 'Maharashtra') | (df_12['Name of State / UT'] == 'Andhra Pradesh') | (df_12['Name of State / UT'] == 'Uttar Pradesh')]
df_filtered.reset_index()
#df_filtered.reset_index(drop = True)
print(df_filtered)
```

```python
sb.set(style="whitegrid")
clarity_ranking = ['Maharashtra','Rajasthan','Tamil Nadu','Kerala','Andhra Pradesh','Delhi','Uttar Pradesh']
sb.boxenplot(x="Name of State / UT", y="Total Confirmed cases (Including 76 foreign Nationals) ",
              color="b", order=clarity_ranking,
              scale="linear", data=df_filtered)
plt.show()
```
